# Bubble sort
<a name="bubble"></a>
## Основная информация
<a name="bubblecode"></a>
* <font size="4">Сортировка пузырьком.</font>
* <font size="4">Простой алгоритм, но неэффективный. Его даже называют настолько плохим, что он не нужен в качестве примера алгоритма при обучении. Именно поэтому мы его разберем!</font>

## Псевдокод
<a name="bubblecode"></a>
<font size="4"><pre>
1  BubbleSort(Array):
2      n = length(Array)
3      <b>repeat</b>
4          swapped = <b>False</b>
5          <b>for</b> i = 1 <b>to></b> n-1 <b>do</b>
6              <b>if</b> Array[i-1] > Array[i] <b>then</b>
7                  swap(Array[i-1], Array[i])
8                  swapped = <b>True</b>
9          n = n - 1
10     <b>until not</b> swapped
</pre></font>

# Selection sort
<a name="selection"></a>
<font size="4">Сортировка выбором - простой принцип: находится наименьший элемент, меняется местами с первым неотсортированным.</font> 

## Псевдокод
<a name="selectioncode"></a>
<font size="4">
<pre>
1  SelectionSort(Array):
2      <b>for</b> i <b>in</b> 0..n-1
3          min = i
4          <b>for</b> j <b>in</b> i+1..n-1
5              <b>if</b> Array[j] < Array[min]
6                  min = j
7          <b>if</b> i != min
8              swap(Array, i, min)
</pre>
</font>

## Анализ сложности
<a name="selectioncomplexity"></a>
<font size="4">
    <ol>
        <li>Сколько всего сравнений выполняется в сортировке?</li>
        <li>Сколько шагов во внешнем цикле?</li>
        <li>Сколько шагов во вложенном цикле?</li>
    </ol>
</font>

### Количество сравнений

<font size="4">
    <br>
- непрозрачно
 $$N_{steps} = \sum_{i=0}^{n-1} \sum_{j=i+1}^{n-1} 1$$ 
- чуть лучше
 $$N_{steps} = \sum_{i=0}^{n-1} n-i-1 $$
- интуитивный вариант
 $$N_{steps} = \sum_{i=1}^{n-1} i $$
- без сумм
 $$ 1+2+...+(n - 1) = \dfrac{n (n-1)} {2} $$
</font>

 
### Как решать рекуррентности?
<font size="4">Вопрос подробно рассматривается в книге "Конкретная математика" Р. Грэхема и Д. Кнута.</font>
 
## Доказательство корректности
<a name="selectionproof"></a>
### Как организовано доказательство
<font size="4">
    Если в алгоритме присутствует цикл (то есть <i>почти</i> всегда)
    <ol>
        <li>Инвариант цикла</li>
        <li>Сохранение инварианта</li>
        <li>Завершение цикла</li>
    </ol> 
</font>

### Наш случай
<font size="4">
    Если в алгоритме присутствует цикл (то есть <i>почти</i> всегда):
    <ol>
        <li><i>Инвариант</i>: начало списка до $ i $-го элемента отсортировано. При инициалиизации (пустой список) верно</li>
        <li><i>Сохранение инварианта</i>: после итерации цикла $ i += 1 $; начало списка отстортировано</li>
        <li><i>Завершение цикла</i>: $ i $ = длине списка, весь список отсортиррован</li>
    </ol> 
</font>


### Определение отсортированного списка

<font size="4">Важно дать формальное определение, потому что оно будет имспользоваться при проверке сохранения <i>инварианта</i>
> Такая перестановка (элементов массива), что для всех $ i $ , где $ i $ - индекс элемента в списке, $ A_i \leq A_{i+1} $
</font>
<img src="files/sorted_array_upd.png">


### Почему так?
<font size="4">
    <ol>
        <li>Пустой список отсортирован по соглашению</li>
        <li>На каждом следующем шаге выбираем из списка минимальный элемент</li>
        <li>Здесь надо бы доказать, что в списке есть минимальный элемент, и мы извлекаем его :)</li>
        <li>Находим минимальный элемент, вставляем в <i>конец</i> отсортированного списка. При этом список остается <i>отсортированным</i>!</li>
        <li>Завершение цикла: мы добираемся до конца списка. Весь список отсортирован</li>
    </ol> 
</font>

# Insertion sort
<a name="insertion"></a>
Сортировка вставкой

## Псевдокод
<a name="insertioncode"></a>

<pre>
1  InsertionSort(Array):
2      <b>for</b> i <b>in</b> 0..n-1
3          x = Array[i]
4          j = i - 1
5          <b>while</b> j >= 0 <b>and</b> Array[j] > x
6              Array[j+1] = Array[j] // "сдвиг" вправо
7              j = j - 1
8          Array[j+1] = x  // Вставка x в отсортированную часть
9          i = i + 1
</pre>

### Отличие от Selection Sort

<font size="4">
<ol>
    <li>Основное отличие - достаточно вего 1го сравнения, когда k+1 элемент > k-го элемента</li>
    <li>selection sort будет "пробегать" все оставшиеся элементы</li>
    <li>selection sort записывает в память $ O(n) $ раз, а insertion sort - $ O(n^2) $
</ol>
    
Также Insertion Sort:
<ol>
    <li>Стабильный</li>
    <li>Online</li>
    <li>In-place</li>    
</ol>
</font>

# Shellsort

<font size="4">
Сортировка Шелла (1959г.) - "улучшение" сортировки вставками / сортировки пузырьком
</font>

## Идея сортировки
<a name="shellidea"></a>
<font size="4">
    <ol>
        <li>Разобьем список на части (каждый $k$-й элемент)</li>
        <li>Отсортируем эти части insertion sort</li>
        <li>Повторим с другим разбиением c меньшим $ k $, если $ k > 1 $ </li>
    </ol>
</font>
